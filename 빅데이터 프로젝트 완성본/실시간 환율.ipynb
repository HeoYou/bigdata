```python
def exchange():
    url='https://www.koreaexim.go.kr/site/program/financial/exchangeJSON?authkey=bsTEjguAL6eQadcdchsjzky29u1Qg7zn&searchdate=20190503&data=AP01'
    response=requests.get(url)
    result=response.text
    data=json.loads(result)
    df=pd.DataFrame(data)
    df=df.iloc[:,0:2]
    df=df.rename(columns={'bkpr':'환율','cur_nm':'화폐 단위'})
    df=df.stack().str.replace(',','').unstack()
    gu=df.groupby('화폐 단위')['환율'].sum()
    gu=gu.astype(int)
    graph=plt.plot(gu,'bs')
    plt.title("실시간 환율 정보")
    plt.xticks(rotation=90)
```
